```python
import pandas as pd
df1=pd.read_csv("D:/HR.csv")
print(df1) #fetching hr data from system drive D
```

```python
X = df1
X =  pd.get_dummies(X)#one hot encoding for sales and salary 
X.head()
Y=X['left'] #decison or output Y as left
X=X.drop(['left'],axis=1) #dropping it from actual set
print(X)
```

```python
from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=scaler.fit_transform(X)
```

```python
X_train, X_test,Y_train, Y_test=train_test_split(X,Y,test_size=0.3, random_state=42) 
#training data 70 percent and test data 30 percent
```

```python
from sklearn.linear_model import LogisticRegression #applying model linear regression
model=LogisticRegression(penalty='l2',C=1)
```

```python
model.fit(X_train,Y_train) #training data
```

```python
result = model.score(X_test, Y_test) #checking results of model by applying test data
print(("Accuracy: %.3f%%") % (result*100.0))
```
